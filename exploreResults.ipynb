```python
import pandas as pd
import os
import plotly.express as px
```

```python
from IPython.display import HTML

def jupyter_print(df,full=False):
    if full== False:    
        return display(HTML(df[:20].to_html()))
    else:
        return display(HTML(df.to_html()))
```

```python
path = 'results/detailed_output'
filename1 = 'detailed_results.csv'
filename2 = 'models_predictions.csv'

models_results = pd.read_csv(os.path.join(path,filename1))
models_pred = pd.read_csv(os.path.join(path,filename2))
```

```python
fig = px.scatter(
    models_pred, 
    x='Speed_Over_Ground_(knots)', 
    y=['y_true', "y_pred_('Random Forest', 'simple')"], 
    hover_data = ['Speed_Over_Ground_(knots)', 
    'Speed_Through_Water_(knots)',
       'Leeway_(°)', 'Wind_Speed_sourceShip(kn)',
       'significant_height_of_combined_wind_waves_and_swell',
       'mean_wave_period', 'True_Wave_Angle_meto_shiRef_(°)',
       'True_Wind_Speed_meto_(kn)', 'True_Wind_Angle_meto_shiRef_(°)',
       'Apparent_Wind_speed_meto_shipRef(kn)',
       'Apparent_Wind_angle_meto_shipRef(°)'],
    labels={'variable': 'Y Columns'},
    width=1000, height=600,
    )

# Update trace to set different colors for each Y column
fig.update_traces(marker=dict(size=10), mode='markers')
```
